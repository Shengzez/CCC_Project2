```python
df[2017] = df.apply(lambda x:(x[" rai_cityadjusted_total_2017_q1"]+x[" rai_cityadjusted_total_2017_q2"]
                    +x[" rai_cityadjusted_total_2017_q3"]+x[" rai_cityadjusted_total_2017_q4"])/4,axis =1)

df[2018] = df.apply(lambda x:(x[" rai_cityadjusted_total_2018_q1"]+x[" rai_cityadjusted_total_2018_q2"]
                    +x[" rai_cityadjusted_total_2018_q3"]+x[" rai_cityadjusted_total_2018_q4"])/4,axis =1)

df[2019] = df.apply(lambda x:(x[" rai_cityadjusted_total_2019_q1"]+x[" rai_cityadjusted_total_2019_q2"]
                    +x[" rai_cityadjusted_total_2019_q3"]+x[" rai_cityadjusted_total_2019_q4"])/4,axis =1)

df[2020] = df.apply(lambda x:(x[" rai_cityadjusted_total_2020_q1"]+x[" rai_cityadjusted_total_2020_q2"]
                    +x[" rai_cityadjusted_total_2020_q3"]+x[" rai_cityadjusted_total_2020_q4"])/4,axis =1)

df[2021] = df.apply(lambda x:(x[" rai_cityadjusted_total_2021_q1"]+x[" rai_cityadjusted_total_2021_q2"])/2,axis =1)


fram = [df[2017],df[2018],df[2019],df[2020],df[2021], df['geography_name']]
result = pd.concat(fram,axis =1)
```
